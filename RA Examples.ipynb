```python
import pandas as pd
import numpy as np
from sklearn import datasets
from RegressionAnalysis import RegressionAnalysis as ra
iris = pd.DataFrame(datasets.load_iris()['data'])
iris.columns = ['s_length', 's_width', 'p_length', 'p_width']
```

# linear-linear 

```python
r = ra(iris, 
       'linear-linear', 
       's_length', 
       ad_channels = ['s_width', 'p_length', 'p_width'])
```

```python
model = r.main_model()
```

```python
cont = r.calc_contribution()
```
